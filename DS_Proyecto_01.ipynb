```python
# Identifica ouliers

#sns.boxplot(data = properati_ds_top, y = "lon")               # 180
#sns.boxplot(data = properati_ds_top, x = "lat")               # 80 y algo
#sns.boxplot(data = properati_ds_top, x = "rooms")             # del 6 al 35 
#sns.boxplot(data = properati_ds_top, x = "bedrooms")          # del 7 al 15
#sns.boxplot(data = properati_ds_top, x = "bathrooms")         # del 4 al 14

#sns.boxplot(data = properati_ds_top, x = "surface_total") # muchos..

properati_ds_top.surface_total.describe()

# Quitar registros sin valor
df_filter1 = properati_ds_top.dropna(subset=['surface_total', 'surface_covered'])
df_filter1.shape

# Quitar registros donde surface_covered sea más grande que surface_total
df_filter2 = df_filter1[df_filter1['surface_total'] > df_filter1['surface_covered']]
df_filter2.shape

# Quitar registros donde surface_total este entre 0 y 20 ya que en mi punto de vista aportan valor propiedades tan pequeñas 
df_filter3 = df_filter2[df_filter2.surface_total >= 21]

sns.boxplot(data = df_filter3, x = "surface_total")
plt.title("Surface Total")
plt.show()

df_filter3.shape
```

**Primer Boxplot 'Surface Total'**

Nos brinda un gráfico imposible de leer y determinar la cantidad de outliers que pueda tener.

A simple vista se pueden ver 2 bastante alejados del conjunto de la derecha, mismos que seran retirados para el siguiente ciclo.

```python
df_21_to_10000 = 0
df_10001_to_20000 = 0
df_20001_to_100000 = 0
df_100001_to_130000 = 0

for value in df_filter3.surface_total:
  if 21 <= value <= 10000:
    df_21_to_10000 += 1
  elif 10001 <= value <= 20000:
    df_10001_to_20000 += 1
  elif 20001 <= value <= 100000:
    df_20001_to_100000 += 1
  elif 100001 <= value <= 130000:
    df_100001_to_130000 += 1

surface_total_ranges = pd.DataFrame({'21_to_10000': df_21_to_10000, '10001_to_20000': df_10001_to_20000, '20001_to_100000': df_20001_to_100000, '100001_to_130000': df_100001_to_130000}, index=['count'])
surface_total_ranges

sns.boxplot(data = df_filter3[(df_filter3.surface_total >= 21) & (df_filter3.surface_total <= 10000)], x = "surface_total")
plt.title("Surface Total")
plt.show()
```

**Segundo Boxplot 'Surface Total'**

Para determinar este gráfico, lo que hice fue hacer rangos para determinar la cantidad de datos que existia en cada uno para poder sacar algún rango con muy poco valores, así reducir outliers y volver a mirar el gráfico sin dicho rango. 

Ahora podemos darnos una idea más cercana de la gran cantidad de outliers con los que cuenta esta variable.

Aún así debere sacar otro ciclo, para lograr ver la concentración de los datos.

```python
df_21_to_2000 = 0
df_2001_to_4000 = 0
df_4001_to_6000 = 0
df_6001_to_8000 = 0
df_8001_to_10000 = 0

for value in df_filter3.surface_total:
  if 21 <= value <= 2000:
    df_21_to_2000 += 1
  elif 2001 <= value <= 4000:
    df_2001_to_4000 += 1
  elif 4001 <= value <= 6000:
    df_4001_to_6000 += 1
  elif 6001 <= value <= 8000:
    df_6001_to_8000 += 1
  elif 8001 <= value <= 10000:
    df_8001_to_10000 += 1

surface_total_ranges2 = pd.DataFrame({'21_to_2000': df_21_to_2000, '2001_to_4000': df_2001_to_4000, '4001_to_6000': df_4001_to_6000, '6001_to_8000': df_6001_to_8000, '8001_to_10000': df_8001_to_10000}, index=['count'])
surface_total_ranges2

sns.boxplot(data = df_filter3[(df_filter3.surface_total >= 21) & (df_filter3.surface_total <= 2000)], x = "surface_total")
plt.title("Surface Total")
plt.show()
```

```python
# Cargar el dataset
#properati = pd.read_csv("./drive/MyDrive/Acámica/Sprints/Sprint 1/DS_Proyecto_01_Datos_Properati.csv")
properati = pd.read_csv("./DS_Proyecto_01_Datos_Properati.csv")
print(f'properati shape: {properati.shape}')

# Filtrar
# 1. Selecciona aquellas propiedades en Capital Federal y cuyo tipo de propiedad es Departamento, PH o Casa.
properati_1 = properati[(properati.l2 == 'Capital Federal') & (properati.property_type.isin(['Departamento', 'PH', 'Casa']))]
print(f'properati_1 shape: {properati_1.shape}')

# 2. Selecciona aquellas propiedades cuya superficie total es menor a 1000 m2 y mayor a 15 m2.
properati_2 = properati_1[(properati_1.surface_total <= 1000) & (properati_1.surface_total >= 15)]
print(f'properati_2 shape: {properati_2.shape}')

# 3. Selecciona aquellas propiedades cuya precio es menor 4000000 dólares.
properati_3 = properati_2[properati_2.price <= 4000000]
print(f'properati_3 shape: {properati_3.shape}')

# 4. Selecciona las columnas rooms, bedrooms, bathrooms, surface_total, surface_covered y price.
properati_4 = properati_3.filter(['rooms', 'bedrooms', 'bathrooms', 'surface_total', 'surface_covered', 'price'])
print(f'properati_4 shape: {properati_4.shape}')

# 5. Descarta aquellas instacias con valores faltantes.
properati_5 = properati_4.dropna()
print(f'properati_5 shape: {properati_5.shape}')
```

```python
corr = properati_5.filter(['rooms', 'bedrooms', 'bathrooms', 'surface_total', 'surface_covered', 'price']).corr()

sns.heatmap(corr, linewidths=.5, cmap="YlGnBu", annot=True)
plt.title("Correlación")
plt.show()
```

```python
x1 = properati_5.filter(['surface_total'])
x2 = properati_5.filter(['bathrooms'])
y = properati_5.filter(['price'])
```

```python
from mpl_toolkits.mplot3d import Axes3D

def plot_3d_scatter(x1, x2, y, ax=None, fig = None):
    if (fig is None) and (ax is None):
        fig = plt.figure(figsize = (8,6))
        ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x1, x2, y, alpha=0.5)

    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.set_zlabel('y')

plot_3d_scatter(x1, x2, y, ax=None, fig = None)
```

```python
from sklearn.model_selection import train_test_split

X = properati_5[['surface_total', 'bathrooms']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

linear_model = LinearRegression()
tree_regressor = DecisionTreeRegressor(max_depth=3, random_state=42)
knn_regressor = KNeighborsRegressor(n_neighbors=5)
```

```python
linear_model.fit(X_train, y_train)
tree_regressor.fit(X_train, y_train)
knn_regressor.fit(X_train, y_train)
```

```python
from matplotlib import cm

def plot_3d_regressor(x1_min, x1_max, x2_min,x2_max, N, regressor, ax=None, fig = None):
    x1 = np.linspace(x1_min,x1_max,N) 
    x2 = np.linspace(x2_min,x2_max,N)
    X1, X2 = np.meshgrid(x1,x2)
    
    y = regressor.predict(np.array([X1.ravel(), X2.ravel()]).T)
    Y = y.reshape(X1.shape)
    
    if (fig is None) and (ax is None):
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
    
    surf = ax.plot_surface(X1, X2, Y, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False, alpha = 0.5)
```

```python
fig = plt.figure(figsize = (12,4))

ax = fig.add_subplot(131, projection='3d')
plt.title('Regresión Lineal')
plot_3d_regressor(x1.surface_total.min(), x1.surface_total.max(), x2.bathrooms.min(), x2.bathrooms.max(), 100, linear_model, ax, fig)
plot_3d_scatter(x1, x2, y, ax, fig)

ax = fig.add_subplot(132, projection='3d')
plt.title('Árbol de Decisión')
plot_3d_regressor(x1.surface_total.min(), x1.surface_total.max(), x2.bathrooms.min(), x2.bathrooms.max(), 100, tree_regressor, ax, fig)
plot_3d_scatter(x1, x2, y, ax, fig)

ax = fig.add_subplot(133, projection='3d')
plt.title('Vecinos más cercanos')
plot_3d_regressor(x1.surface_total.min(), x1.surface_total.max(), x2.bathrooms.min(), x2.bathrooms.max(), 100, knn_regressor, ax, fig)
plot_3d_scatter(x1, x2, y, ax, fig)

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
import seaborn as sns
modelos = ['Regresión lineal', 'Árbol de Decisión', 'Vecinos más cercanos']

for i, model in enumerate([linear_model, tree_regressor, knn_regressor]):
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    if modelos[i] == 'Árbol de Decisión':
      y_train_pred = pd.DataFrame(y_train_pred, columns=['price'])
      y_test_pred = pd.DataFrame(y_test_pred, columns=['price'])
    
    dif_train = y_train - y_train_pred
    dif_test = y_test - y_test_pred
        
    dif = pd.DataFrame({'Diff_Train': dif_train.price, 'Diff_Test': dif_test.price})
    
    plt.figure(figsize = (8,4))

    plt.subplot(1,2,1)
    sns.histplot(dif, bins = 20, label = 'train', x='Diff_Train', kde=True)
    sns.histplot(dif, bins = 20, label = 'test', x='Diff_Test', color='orange', kde=True)
    plt.xlabel('errores')
    plt.legend()

    ax = plt.subplot(1,2,2)
    ax.scatter(y_test,y_test_pred, s =2)    
    lims = [
    np.min([ax.get_xlim(), ax.get_ylim()]),  # min of both axes
    np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes]
    ]
    
    ax.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
    plt.xlabel('y (test)')
    plt.ylabel('y_pred (test)')
    
    plt.tight_layout()

    print()
    print()
    print(f'Modelo: {modelos[i]}')
    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
    print(f'Raíz del error cuadrático medio en Train: {rmse_train}')
    print(f'Raíz del error cuadrático medio en Test: {rmse_test}')
    print()

    plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = properati_5[['rooms', 'bedrooms', 'bathrooms', 'surface_total', 'surface_covered']]
y = properati_5.filter(['price'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
print('Árbol de Decisión - 5 variables\n')

array_rsme_tree = pd.DataFrame(columns = ['Train', 'Test', 'max_depth'])

for n in range(1, 50, 1):
    tree_regressor_ciclo = DecisionTreeRegressor(max_depth=n, random_state=42)
    
    tree_regressor_ciclo.fit(X_train, y_train)
    
    y_train_pred_ciclo = tree_regressor_ciclo.predict(X_train)
    y_test_pred_ciclo = tree_regressor_ciclo.predict(X_test)
    
    rmse_train_ciclo = np.sqrt(mean_squared_error(y_train, y_train_pred_ciclo))
    rmse_test_ciclo = np.sqrt(mean_squared_error(y_test, y_test_pred_ciclo))
    
    array_rsme_tree = array_rsme_tree.append({'Train': rmse_train_ciclo, 'Test': rmse_test_ciclo, 'max_depth': n}, ignore_index=True)

min_train = array_rsme_tree.Train.min()
maxdepth_min_train = array_rsme_tree[array_rsme_tree.Train == min_train].reset_index()
maxdepth_min_train = maxdepth_min_train.loc[0,'max_depth']
min_test = array_rsme_tree.Test.min()
maxdepth_min_test = array_rsme_tree[array_rsme_tree.Test == min_test].reset_index()
maxdepth_min_test = maxdepth_min_test.loc[0,'max_depth']

print(f'El RSME más bajo en Train es: {min_train} -> con un K de {maxdepth_min_train}')
print(f'El RSME más bajo en Test es: {min_test} -> con un K de {maxdepth_min_test}')


sns.lineplot(data=array_rsme_tree, x="max_depth", y="Train", label='Train')
sns.lineplot(data=array_rsme_tree, x="max_depth", y="Test", label='Test')

plt.xlabel('max_depth')
plt.ylabel('RSME')

plt.title('RSME DecisionTreeRegressor: Train vs Test')
plt.show()
```

```python
print('Vecinos más Cercanos\n')

array_rsme_knn = pd.DataFrame(columns = ['Train', 'Test', 'max_depth'])

for n in range(1, 100, 1):
    knn_regressor = KNeighborsRegressor(n_neighbors=n)
    
    knn_regressor.fit(X_train, y_train)
    
    y_train_pred_ciclo = knn_regressor.predict(X_train)
    y_test_pred_ciclo = knn_regressor.predict(X_test)
    
    rmse_train_ciclo = np.sqrt(mean_squared_error(y_train, y_train_pred_ciclo))
    rmse_test_ciclo = np.sqrt(mean_squared_error(y_test, y_test_pred_ciclo))
    
    array_rsme_knn = array_rsme_knn.append({'Train': rmse_train_ciclo, 'Test': rmse_test_ciclo, 'max_depth': n}, ignore_index=True)

min_train = array_rsme_knn.Train.min()
maxdepth_min_train = array_rsme_knn[array_rsme_knn.Train == min_train].reset_index()
maxdepth_min_train = maxdepth_min_train.loc[0,'max_depth']
min_test = array_rsme_knn.Test.min()
maxdepth_min_test = array_rsme_knn[array_rsme_knn.Test == min_test].reset_index()
maxdepth_min_test = maxdepth_min_test.loc[0,'max_depth']

print(f'El RSME más bajo en Train es: {min_train} -> con un K de {maxdepth_min_train}')
print(f'El RSME más bajo en Test es: {min_test} -> con un K de {maxdepth_min_test}')


sns.lineplot(data=array_rsme_knn, x="max_depth", y="Train", label='Train')
sns.lineplot(data=array_rsme_knn, x="max_depth", y="Test", label='Test')

plt.xlabel('max_depth')
plt.ylabel('RSME')

plt.title('RSME KNeighborsRegressor: Train vs Test')
plt.show()
```
