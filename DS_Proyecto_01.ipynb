**Departamento**

En un departamento, realmente estoy de acuerdo que no importa que tengas en el departamento, importa el tamaño. Pero no deja de ser importante todo lo demas, por ello la correlación de las demás variables es muy pareja.

**PH**

En un PH, al ser algo que puede ser usado de muchas formas lo más importante es el tamaño (aun que pense que seria más importante que la del departamento), pero se denota lo que digo al ver como disminuyo el porcentaje en las demás variables.

**Casa**

Aqui si esta muy extraño, si es muy importante el tamaño como en un departamento, pero que los baños no tenga casi ningun impacto no se me hace coerente ya que una casa sin baños... si deberia de afectar al precio.
También aqui se ve nota la importancia de los cuartos, por que al menos en México hay casas que lo primero que te dicen es ('número de cuartos', antes que alguna otra información) y esto hablando de arquitectura tiene sentido porque la distribución de una casa radica en la cantidad de cuartos dentro de la superficie cubierta.

**Importante:** para asegurarnos que trabajes con un dataset apropiados, debes volver a cargar los datos y realizar el siguiente filtrado:

1. Selecciona aquellas propiedades en Capital Federal y cuyo tipo de propiedad es Departamento, PH o Casa.
1. Selecciona aquellas propiedades cuya superficie total es menor a 1000 m2 y mayor a 15 m2.
1. Selecciona aquellas propiedades cuya precio es menor 4000000 dólares.
1. Selecciona las columnas `rooms`, `bedrooms`, `bathrooms`, `surface_total`, `surface_covered` y `price`.
1. Descarta aquellas instacias con valores faltantes.

**Checkpoint:** deberías obtener un dataset con 81019 instacias y 6 columnas.

```python
# Cargar el dataset
#properati = pd.read_csv("./drive/MyDrive/Acámica/Sprints/Sprint 1/DS_Proyecto_01_Datos_Properati.csv")
properati = pd.read_csv("./DS_Proyecto_01_Datos_Properati.csv")
print(f'properati shape: {properati.shape}')

# Filtrar
# 1. Selecciona aquellas propiedades en Capital Federal y cuyo tipo de propiedad es Departamento, PH o Casa.
properati_1 = properati[(properati.l2 == 'Capital Federal') & (properati.property_type.isin(['Departamento', 'PH', 'Casa']))]
print(f'properati_1 shape: {properati_1.shape}')

# 2. Selecciona aquellas propiedades cuya superficie total es menor a 1000 m2 y mayor a 15 m2.
properati_2 = properati_1[(properati_1.surface_total <= 1000) & (properati_1.surface_total >= 15)]
print(f'properati_2 shape: {properati_2.shape}')

# 3. Selecciona aquellas propiedades cuya precio es menor 4000000 dólares.
properati_3 = properati_2[properati_2.price <= 4000000]
print(f'properati_3 shape: {properati_3.shape}')

# 4. Selecciona las columnas rooms, bedrooms, bathrooms, surface_total, surface_covered y price.
properati_4 = properati_3.filter(['rooms', 'bedrooms', 'bathrooms', 'surface_total', 'surface_covered', 'price'])
print(f'properati_4 shape: {properati_4.shape}')

# 5. Descarta aquellas instacias con valores faltantes.
properati_5 = properati_4.dropna()
print(f'properati_5 shape: {properati_5.shape}')
```

```python
corr = properati_5.filter(['rooms', 'bedrooms', 'bathrooms', 'surface_total', 'surface_covered', 'price']).corr()

sns.heatmap(corr, linewidths=.5, cmap="YlGnBu", annot=True)
plt.title("Correlación")
plt.show()
```

```python
x1 = properati_5.filter(['surface_total'])
x2 = properati_5.filter(['bathrooms'])
y = properati_5.filter(['price'])
```

```python
from mpl_toolkits.mplot3d import Axes3D

def plot_3d_scatter(x1, x2, y, ax=None, fig = None):
    if (fig is None) and (ax is None):
        fig = plt.figure(figsize = (8,6))
        ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x1, x2, y, alpha=0.5)

    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.set_zlabel('y')

plot_3d_scatter(x1, x2, y, ax=None, fig = None)
#plot_3d_scatter(x2, x1, y, ax=None, fig = None)
```

```python
from sklearn.model_selection import train_test_split

#X = properati_5.filter(['surface_total', 'bathrooms'])
X = properati_5[['surface_total', 'bathrooms']]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

linear_model = LinearRegression()
tree_regressor = DecisionTreeRegressor(max_depth=3, random_state=42)
knn_regressor = KNeighborsRegressor(n_neighbors=5)
```

```python
linear_model.fit(X_train, y_train)
tree_regressor.fit(X_train, y_train)
knn_regressor.fit(X_train, y_train)
```

```python
from matplotlib import cm

def plot_3d_regressor(x1_min, x1_max, x2_min,x2_max, N, regressor, ax=None, fig = None):
    x1 = np.linspace(x1_min,x1_max,N) 
    x2 = np.linspace(x2_min,x2_max,N)
    X1, X2 = np.meshgrid(x1,x2)
    
    y = regressor.predict(np.array([X1.ravel(), X2.ravel()]).T)
    Y = y.reshape(X1.shape)
    
    if (fig is None) and (ax is None):
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
    
    surf = ax.plot_surface(X1, X2, Y, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False, alpha = 0.5)
```

```python
fig = plt.figure(figsize = (12,4))

ax = fig.add_subplot(131, projection='3d')
plt.title('Regresión Lineal')
plot_3d_regressor(x1.surface_total.min(), x1.surface_total.max(), x2.bathrooms.min(), x2.bathrooms.max(), 100, linear_model, ax, fig)
plot_3d_scatter(x1, x2, y, ax, fig)

ax = fig.add_subplot(132, projection='3d')
plt.title('Árbol de Decisión')
plot_3d_regressor(x1.surface_total.min(), x1.surface_total.max(), x2.bathrooms.min(), x2.bathrooms.max(), 100, tree_regressor, ax, fig)
plot_3d_scatter(x1, x2, y, ax, fig)

ax = fig.add_subplot(133, projection='3d')
plt.title('Vecinos más cercanos')
plot_3d_regressor(x1.surface_total.min(), x1.surface_total.max(), x2.bathrooms.min(), x2.bathrooms.max(), 100, knn_regressor, ax, fig)
plot_3d_scatter(x1, x2, y, ax, fig)

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
import seaborn as sns
modelos = ['Regresión lineal', 'Árbol de Decisión', 'Vecinos más cercanos']

for i, model in enumerate([linear_model, tree_regressor, knn_regressor]):
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    if modelos[i] == 'Árbol de Decisión':
      y_train_pred = pd.DataFrame(y_train_pred, columns=['price'])
      y_test_pred = pd.DataFrame(y_test_pred, columns=['price'])
    
    plt.figure(figsize = (8,4))

    plt.subplot(1,2,1)
    sns.distplot(y_train - y_train_pred, bins = 20, label = 'train')
    sns.distplot(y_test - y_test_pred, bins = 20, label = 'test')
    plt.xlabel('errores')
    plt.legend()

    ax = plt.subplot(1,2,2)
    ax.scatter(y_test,y_test_pred, s =2)    
    lims = [
    np.min([ax.get_xlim(), ax.get_ylim()]),  # min of both axes
    np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes]
    ]
    
    ax.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
    plt.xlabel('y (test)')
    plt.ylabel('y_pred (test)')
    
    plt.tight_layout()

    print()
    print()
    print(f'Modelo: {modelos[i]}')
    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
    print(f'Raíz del error cuadrático medio en Train: {rmse_train}')
    print(f'Raíz del error cuadrático medio en Test: {rmse_test}')
    print()

    plt.show()
```
